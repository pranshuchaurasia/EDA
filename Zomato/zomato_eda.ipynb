```python
df_country=pd.read_excel(r'Country-Code.xlsx')
```

```python
df_country.columns
```

```python
df_country.head()
```

```python
final_df=pd.merge(df,df_country,on='Country Code',how='left')
```

```python
final_df.head()
```

```python
final_df.dtypes
```

```python
final_df.Country.value_counts()
```

```python
country_names=final_df.Country.value_counts().index
```

```python
final_df.Country.value_counts().values
```

```python
country_val=final_df.Country.value_counts().values
```

```python
# Pie Chart
plt.pie(country_val,labels=country_names)
```

```python
## Top 3 countries
plt.pie(country_val[:3],labels=country_names[:3])
```

```python
plt.pie(country_val[:3],labels=country_names[:3],autopct="%1.2f%%")
```

Observation:</br>
Zomato has maximum user base in India > UK > USA.

```python
ratings=final_df.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size().reset_index().rename(columns={0:'Rating Count'})
```

```python
ratings.head()
```

```python
import matplotlib
matplotlib.rcParams['figure.figsize']=(12,6)
sns.barplot(x="Aggregate rating",y="Rating Count",data=ratings)
```

```python
sns.barplot(x="Aggregate rating",y="Rating Count",data=ratings,hue='Rating color',palette=['white','red','orange','yellow','green','green'])
```

```python
sns.barplot(x="Aggregate rating",y="Rating Count",data=ratings,hue='Rating color',palette=['blue','red','orange','yellow','green','green'])
```

```python
###Count Plot
sns.countplot(x="Rating color",data=ratings,hue='Rating color',palette=['blue','red','orange','yellow','green','green'])
```
